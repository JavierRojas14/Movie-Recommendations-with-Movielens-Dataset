```python
import pandas as pd
```

```python
links = pd.read_csv('links.csv')
movies = pd.read_csv('movies.csv', index_col = 'movieId')
ratings = pd.read_csv('ratings.csv')
tags = pd.read_csv('tags.csv')
```

```python
# EPXLORACIÓN DE DATOS
# Primero se verán las películas que ha rateado un usuario. Quería sortear los datos, pero 
# ya estaban sorteados.

mask_usuario = ratings.userId == 1
ratings[mask_usuario].rating.value_counts()

# Con este mask, se pueden ver cuantas calificaciones ha hecho el usuario.
```

```python
# Ahora, se seleccionaran sólo las películas del usuario que tengan 5 estrellas 
mask_usuario_punt = (ratings.rating == 5) & (ratings.userId == 1)
peliculas_5_rating_user_1 = ratings[mask_usuario_punt]
peliculas_5_rating_user_1
```

```python
# Ahora, se deben agregar las películas, junto a sus categorías, para saber cuales son las 
# categorías que más le gustan al usuario. 

# Para esto, se debe iterar por cada fila de la columna movieId, y extraer los datos desde 
# el df movies

nombres_peliculas = [movies.loc[id].title for id in peliculas_5_rating_user_1.movieId]
generos_peliculas = [movies.loc[id].genres for id in peliculas_5_rating_user_1.movieId]

# Estos datos se deben agregar a columnas en el data frame anterior
peliculas_5_rating_user_1['title'] = nombres_peliculas
peliculas_5_rating_user_1['genres'] = generos_peliculas

peliculas_5_rating_user_1
```
